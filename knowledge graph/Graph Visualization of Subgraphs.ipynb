```python
import numpy as np
import pandas as pd
import math
import os
import networkx as nx
import matplotlib.pyplot as plt
```

```python
save_dir = "Saved Data/G_withDOI_v0_withNodeSize/"

G = nx.read_graphml(save_dir + "Graph.graphml")
```

```python
df = pd.read_csv("../Data (Reformatted)/df_all.csv")
```

```python
all_mat = set(df[(df["mat"]!="unknown")]["mat"].dropna())


all_app = set(df[(df["app"]!="unknown")]["app"].dropna())


all_prod = set(df[(df["prod"]!="unknown")]["prod"].dropna())
```

```python
def map_type_to_color(node_types):
    mapping = {
        'DOI': 'blue',
        'material': 'orange',
        'application': 'yellow',
        'product': 'green'
    }

    return [mapping[str(node_type)] for node_type in node_types]
```

# MAT centric

## coal fly ash

```python
nodes_of_interest = ["coal fly ash"] + list(all_app)

subgraph = G.subgraph(nodes_of_interest)

degrees = dict(nx.degree(subgraph, weight="weight"))

degrees
```

```python
MAT = "coal fly ash"
APPs = ['supplementary cementitious material','geopolymer', 'fine aggregate', 'coarse aggregate', 'clinker feedstock', 
       'reinforced fibre','lightweight aggregate', 'superplasticizer', 'filler']

nodes_of_interest = [MAT] + list(APPs)

subgraph = G.subgraph(nodes_of_interest)

node_types = list(nx.get_node_attributes(subgraph, "node_type").values())
node_colors = map_type_to_color(node_types)

degrees = dict(nx.degree(subgraph, weight="weight"))
node_sizes = [5*item[1] for item in sorted(degrees.items(), key=lambda x: list(subgraph.nodes()).index(x[0]))]

"""
labels = {}
for node in nodes_of_interest:
    if node == "supplementary cementitious material":
        labels[node] = "supplementary\n cementitious\n material"
    else:
        labels[node] = node
"""

labels = {}
for node in nodes_of_interest:
    if node == MAT:
        labels[node] = node
        continue
        
    strs = node.split(" ")
    if len(strs) == 1:
        labels[node] = node
    else:
        strs_new = [w+"\n" for w in strs[:-1]] + [strs[-1]]
        labels[node] = ''.join(strs_new)


#Plotting Graph
pos = nx.shell_layout(subgraph, nlist=[[MAT], list(APPs)])
nx.draw(subgraph, pos, with_labels=True, node_color=node_colors, node_size=node_sizes,
       font_size=11, labels=labels)

#plt.savefig("../Figs/Graph/G coal fly ash.jpg", bbox_inches='tight',dpi=100)
```
